```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
import scipy.cluster.hierarchy as shc
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA as PCA
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('Leukemia.csv')
```

```python
X = data.drop('type', axis = "columns")
Y = data['type']
feature_names = X.columns
```

Podatke svkakao moramo prvo normalizovati

### Normalizacija


```python
scaler = MinMaxScaler()
x = pd.DataFrame(scaler.fit_transform(X), columns = feature_names)
```

```python
plt.figure(figsize = (8, 8))
stablo = shc.dendrogram((shc.linkage(x, method = 'ward')))
plt.show()
```

```python
Y.replace("AML", 0, inplace = True)
Y.replace("Bone_Marrow", 1, inplace = True)
Y.replace("Bone_Marrow_CD34", 2, inplace = True)
Y.replace("PB", 3, inplace = True)
Y.replace("PBSC_CD34", 4, inplace = True)
Y.unique()
```

- 0 -> AML
- 1 -> Bone_Marrow
- 2 -> Bone_Marrow_CD34
- 3 -> PB
- 4 -> PBSC_CD34

```python
pca = PCA()
x_pca = pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_[0:2].sum()
```

### Sakupljajuce hijerarhijsko klasterovanje

```python
linkages = ['average', 'complete', 'ward', 'single']
ks = [i for i in range(2, 14)]
sse = []
silh = []
fig, axs = plt.subplots(12, 4, figsize=(20, 60))
fig.suptitle("")
i = 0
ax = axs.flatten()
best_model = None
max_score = float('-inf')


for k in ks:
    for l in linkages:
        hijerarhijskoA = AgglomerativeClustering(n_clusters = k, linkage = l)
        hijerarhijskoA.fit(x)
        
        score = silhouette_score(x, hijerarhijskoA.labels_)
        ax[i].scatter(x_pca[:, 0], x_pca[:, 1], c=hijerarhijskoA.labels_)
        ax[i].set_title(f"broj_klastera={k}, linkage={l}, score={round(score, 5)}")
        ax[i].set_xlabel("pca_{1}")
        ax[i].set_ylabel("pca_{2}")
        i = i + 1
        
        silh.append(silhouette_score(x_pca, hijerarhijskoA.labels_))
        
        if score > max_score and k != 2:
            max_score = score
            best_model = hijerarhijskoA
        
        
plt.tight_layout()
plt.show()
```

```python
print(best_model.n_clusters, max_score, best_model.linkage)
```

```python
plt.figure(figsize = (8, 8))
stablo = shc.dendrogram((shc.linkage(x, method = 'average')))
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(ks, silh, marker = 'p')
ax.set_xlabel("br klastera")
ax.set_ylabel("silueta")
plt.show()
```

Dakle, za sakupljajuce hijerarhijsko klasterovanje izgleda da najbolji skor siluete imamo u 2 klastera, ali posto zelimo neki veci broj, izabracemo 10.

```python
hijerarhijskoA2 = AgglomerativeClustering(n_clusters = 2, memory = 'memory')
hijerarhijskoA2.fit(x)
```

```python
silhouette_score(x, hijerarhijskoA2.labels_)
```
